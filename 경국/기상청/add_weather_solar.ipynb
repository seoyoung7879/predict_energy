```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rc
import platform

# ✅ 한글 폰트 설정
if platform.system() == 'Windows':
    plt.rc('font', family='Malgun Gothic')
elif platform.system() == 'Darwin':
    plt.rc('font', family='AppleGothic')
else:
    plt.rc('font', family='NanumGothic')

plt.rcParams['axes.unicode_minus'] = False

# ✅ 1. CSV 파일 불러오기
file_path = '한국서부발전_시간별발전량_태양광_20231231_준공연도.csv'
df = pd.read_csv(file_path, encoding='utf-8-sig')

df = df.rename(columns={
    '발전기명': '발전구분',
    '날짜': '일자',
    '해당 시간 발전량(kWh)': '발전량(kWh)'
})
df['시간'] = df['시간'].str.replace('시', '', regex=False).astype(int)

# ✅ 2. 날짜 변환 및 5일 단위 그룹화용 컬럼 생성
df['일자'] = pd.to_datetime(df['일자'])
df['기간'] = (df['일자'].dt.dayofyear // 5).astype(int)  # 5일 단위 bin
df['기간'] = df['일자'].dt.year.astype(str) + '-P' + df['기간'].astype(str).str.zfill(2)  # e.g., '2023-P01'

# ✅ 3. 발전소 컬럼 생성
df['발전소'] = df['발전구분']

# ✅ 4. 발전량 기반 5일 단위 평균
period_avg = df.groupby(['기간', '발전소'])['발전량(kWh)'].mean().reset_index()

# ✅ 5. 설비이용률 계산 (kWh / (MW * 1000))
df['설비이용률'] = df['발전량(kWh)'] / (df['설비용량(MW)'] * 1000)
period_util = df.groupby(['기간', '발전소'])['설비이용률'].mean().reset_index()

# ✅ 6. 그래프 1: 5일 단위 평균 발전량
plt.figure(figsize=(18, 8))
for name, group in period_avg.groupby('발전소'):
    plt.plot(group['기간'], group['발전량(kWh)'], marker='o', label=name)

plt.title('5일 단위 · 발전소별 평균 발전량 (kWh)', fontsize=14)
plt.xlabel('기간 (5일 단위)', fontsize=12)
plt.ylabel('평균 발전량 (kWh)', fontsize=12)
plt.xticks(rotation=90)
plt.grid(True)
plt.legend(title='발전소', bbox_to_anchor=(1.01, 1), loc='upper left')
plt.tight_layout()
plt.show()

# ✅ 7. 그래프 2: 5일 단위 평균 설비이용률
plt.figure(figsize=(18, 8))
for name, group in period_util.groupby('발전소'):
    plt.plot(group['기간'], group['설비이용률'], marker='o', label=name)

plt.title('5일 단위 · 발전소별 평균 설비이용률 (kWh/kW)', fontsize=14)
plt.xlabel('기간 (5일 단위)', fontsize=12)
plt.ylabel('평균 설비이용률', fontsize=12)
plt.xticks(rotation=90)
plt.grid(True)
plt.legend(title='발전소', bbox_to_anchor=(1.01, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

## 🌦 기상청 예보 데이터 컬럼 설명

| 컬럼명 | 의미 | 단위 또는 값 |
|--------|------|---------------|
| `fcstDate` | 예보 날짜 | 예: `20250627` → 2025년 6월 27일 |
| `fcstTime` | 예보 시각 | 예: `0600` → 오전 6시 |
| `TMP` | 기온 | ℃ (섭씨 온도) |
| `SKY` | 하늘 상태 | 1: 맑음, 3: 구름 많음, 4: 흐림 |
| `PTY` | 강수 형태 | 0: 없음, 1: 비, 2: 비/눈, 3: 눈, 4: 소나기 등 |
| `POP` | 강수 확률 | % |
| `PCP` | 1시간 강수량 | mm 또는 `"강수없음"` |
| `REH` | 습도 | % |
| `WSD` | 풍속 | m/s (미터/초) |
| `SNO` | 1시간 적설량 | cm 또는 `"적설없음"` |
| `TMN` | 일 최저기온 | ℃ (해당 날짜 전체의 최저 예상 기온) |
| `TMX` | 일 최고기온 | ℃ (해당 날짜 전체의 최고 예상 기온) |
| `UUU` | 동서 바람 성분 | m/s (음수: 서풍, 양수: 동풍) |
| `VVV` | 남북 바람 성분 | m/s (음수: 북풍, 양수: 남풍) |
| `VEC` | 풍향 | 도 (°), 0: 북, 90: 동, 180: 남, 270: 서 |
| `WAV` | 파고 (해상용) | m (육상 예보에서는 -999로 무효 값 처리됨) |

---

## 💡 추가 설명

- `UUU`, `VVV`는 바람의 벡터 성분입니다.  
  → 풍속(`WSD`)과 풍향(`VEC`)으로부터 계산 가능  
  → `풍속^2 = UUU^2 + VVV^2`, `풍향 = atan2(UUU, VVV)`

- `TMX`, `TMN`은 하루에 한 번만 나타나며, 그 날 전체를 대표하는 예보입니다.

- `PCP`, `SNO`는 `"강수없음"`, `"적설없음"`처럼 문자열이 포함되어 있어 수치 분석 시 처리 필요합니다.

- `WAV`는 해양 예보 전용 항목이며, 육상에서는 대부분 `-999`로 나타나며 무시해도 무방합니다.

```python
import re

def clean_precip(val):
    # NaN이면 그대로 반환
    if pd.isna(val):
        return val
    val = str(val)
    
    # 강수없음, 적설없음 → 0
    if '없음' in val:
        return 0.0
    
    # 1mm 미만 → 관례적으로 0.5로 처리하거나 0.0으로 처리 가능
    if '미만' in val:
        return 0.5  # 또는 0.0으로 바꿔도 됨
    
    # 숫자만 추출 (예: '2.5mm' → 2.5)
    numeric_part = re.findall(r"[\d.]+", val)
    return float(numeric_part[0]) if numeric_part else 0.0

# 적용
pivot_df['강수량(mm)'] = pivot_df['강수량(mm)'].apply(clean_precip)
pivot_df['적설(cm)'] = pivot_df['적설(cm)'].apply(clean_precip)
```

```python
# 현재 컬럼들 확인
cols = df_weather.columns.tolist()

# 기준이 되는 컬럼 이름 (예: '지점' 또는 '지점명')
key_col = '지점명'  # 또는 '지점'으로 바꿔도 됨

# '지점명' 다음에 '일자', '시간'을 삽입
if key_col in cols:
    idx = cols.index(key_col)
    # 기존에 '일자', '시간'이 이미 들어있다면 먼저 제거
    for c in ['일자', '시간']:
        if c in cols:
            cols.remove(c)
    # 삽입
    new_order = cols[:idx+1] + ['일자', '시간'] + cols[idx+1:]
    df_weather = df_weather[new_order]
else:
    print(f"⚠️ 컬럼 '{key_col}'이 존재하지 않습니다.")

df_weather.head(5)
```

```python
def map_cloud_to_sky(cloud):
    if pd.isna(cloud):
        return None
    try:
        c = float(cloud)
    except:
        return None
    if c <= 2:
        return 1
    elif c <= 7:
        return 3
    else:
        return 4
    
df_weather = df_weather.rename(columns={'전운량(10분위)': '하늘상태'})

# 전운량 컬럼이 '전운량(10분위)'라는 이름일 경우
df_weather['하늘상태'] = df_weather['하늘상태'].apply(map_cloud_to_sky)


df_weather.head(5)
```

### 발전량 데이터 + 기상청 데이터

```python
# 3. 발전소 → 지점명 매핑
plant_to_location = {
    '(군산)영암F1태양광': '목포',
    '(군산)삼랑진태양광': '양산시',
    '태안#9,10 수상태양광': '서산',
}


# 4. 지점명 컬럼 추가
df_energy['지점명'] = df_energy['발전구분'].map(plant_to_location)

df_energy['지점명'] = df_energy['지점명'].astype(str).str.strip()
df_weather['지점명'] = df_weather['지점명'].astype(str).str.strip()

# 5. 날짜 형식 맞추기 (둘 다 'YYYY-MM-DD' 형태여야 병합됨)
df_energy['일자'] = pd.to_datetime(df_energy['일자']).dt.strftime('%Y-%m-%d')
df_weather['일자'] = pd.to_datetime(df_weather['일자']).dt.strftime('%Y-%m-%d')

df_energy['시간'] = df_energy['시간'].astype(int)
df_weather['시간'] = df_weather['시간'].astype(int)

# 6. 병합: 지점명 + 일자 + 시간 기준
df_merged = pd.merge(df_energy, df_weather, on=['지점명', '일자', '시간'], how='left')

# 7. 결과 확인 (앞부분 출력)
print(df_merged.head())

# 8. 결과 저장
df_merged.to_csv('df_merged_solar.csv', index=False, encoding='utf-8-sig')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import math

# ▶ 전체 그래프를 하나의 이미지로 저장할 설정
cols = 2
rows = math.ceil(len(plants) / cols)
fig, axes = plt.subplots(rows, cols, figsize=(16, 6 * rows))
axes = axes.flatten()


# ✅ 1. 날짜 및 기간 처리
df_merged['일자'] = pd.to_datetime(df_merged['일자'], errors='coerce')
df_merged['기간'] = (df_merged['일자'].dt.dayofyear // 5).astype(int)
df_merged['기간'] = df_merged['일자'].dt.year.astype(str) + '-P' + df_merged['기간'].astype(str).str.zfill(2)

# ✅ 2. 효율 계산
df_merged['효율'] = df_merged['발전량(kWh)'] / df_merged['설비용량(MW)'] / 1000  # 단위: MWh/MW

# ✅ 3. 결측값 제거
df_merged = df_merged.dropna(subset=['효율', '일사(MJ/m2)', '기간', '발전구분'])

# ✅ 4. y축 범위 설정
풍속_min, 풍속_max = 0, 2
효율_min, 효율_max = 0, 0.3

# ✅ 5. 발전소 목록
plants = df_merged['발전구분'].unique()

for i, plant in enumerate(plants):
    sub = df_merged[df_merged['발전구분'] == plant]
    if sub.empty:
        continue

    period_avg = sub.groupby('기간')[['일사(MJ/m2)', '효율']].mean().reset_index()
    period_avg = period_avg.sort_values('기간').reset_index(drop=True)

    ax1 = axes[i]
    ax2 = ax1.twinx()

    ax1.plot(period_avg.index, period_avg['일사(MJ/m2)'], 'g-')
    ax2.plot(period_avg.index, period_avg['효율'], 'b-')

    tick_interval = 10
    ticks = list(range(0, len(period_avg), tick_interval))
    tick_labels = period_avg['기간'].iloc[ticks]
    ax1.set_xticks(ticks)
    ax1.set_xticklabels(tick_labels, rotation=90)

    ax1.set_xlabel('기간 (5일 단위)')
    ax1.set_ylabel('일사(MJ/m2)', color='g')
    ax2.set_ylabel('발전 효율 (MWh/MW)', color='b')
    ax1.set_ylim(풍속_min, 풍속_max)
    ax2.set_ylim(효율_min, 효율_max)
    ax1.set_title(plant)

# ▶ 남는 subplot 비우고 저장
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.savefig('solar_일사량-발전효율_ALL.png', dpi=300)
plt.show()

```
