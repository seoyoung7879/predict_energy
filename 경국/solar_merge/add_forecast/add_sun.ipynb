```python
import pandas as pd
import pvlib

# 1) 위도·경도 정보 (실제 발전소명, 위도/경도 입력 필요)
location_info = {
    '감우리': (37.0525, 128.0028),
    '남제주소내': (33.4843, 126.4979),
    '무릉리': (37.4477, 129.1686),
    '부산복합자재창고': (35.1136, 129.0401),
    '부산본부': (35.1796, 129.0756),
    '부산수처리장': (35.1011, 129.0302),
    '부산신항': (35.0957, 128.8192),
    '부산운동장': (35.1535, 129.0595),
    '삼척소내': (37.4477, 129.1686),
    '송당리': (33.4841, 126.7956),
    '신인천 북측부지': (37.4553, 126.6169),
    '신인천소내': (37.4553, 126.6169),
    '신인천전망대': (37.4562, 126.6169),
    '신인천해수구취수구': (37.4562, 126.6169),
    '신인천 주차장': (37.4553, 126.6169),
    '신풍리': (37.1855, 128.2333),
    '영월본부': (37.1832, 128.4617),
    '영월철도부지': (37.1832, 128.4617),
    '용수리': (37.2013, 128.2731),
    '위미2리': (33.2369, 126.6580),
    '이천D(백사면B)': (37.3395, 127.4913),
    '이천시 백사면A': (37.3395, 127.4913),
    '익산 다송리': (35.9682, 126.9973),
    '인천수산정수장': (37.4866, 126.6305),
    '하동공설운동장': (35.0671, 127.7511),
    '하동변전소': (35.0671, 127.7511),
    '하동보건소': (35.0671, 127.7511),
    '하동본부': (35.0671, 127.7511),
    '하동정수장': (35.0671, 127.7511),
    '하동하수처리장': (35.0671, 127.7511),
    '화촌주민참여형': (37.7686, 127.8480),
    '당진자재창고태양광': (36.8935, 126.6466),
    '당진태양광': (36.8935, 126.6466),
    '당진화력수상태양광': (36.8935, 126.6466),
    '울산태양광#1': (35.5384, 129.3114),
    '광양항광양냉장태양광': (34.9276, 127.7607),
    '광양항어울림태양광': (34.9276, 127.7607),
    '수원환경사업소태양광': (37.2697, 127.0095),
    '농촌공진도태양광': (34.4827, 126.1122),
}


# 2) 시간 컬럼 처리 (예시: 'year', 'month', 'day', 'hour' 컬럼이 있다고 가정)
df_1['year'] = df_1['year'].astype(int)
df_1['month'] = df_1['month'].astype(int)
df_1['day'] = df_1['day'].astype(int)
if 'hour' not in df_1.columns:
    df_1['hour'] = 12  # 시간 정보 없으면 정오(12시)로 설정

df_1['일시'] = pd.to_datetime(
    df_1[['year', 'month', 'day', 'hour']]
).dt.tz_localize('Asia/Seoul')

# 3) 태양 위치 계산 함수
def compute_solar_pos(row):
    plant = row['발전구분']
    dt = row['일시']
    if plant not in location_info or pd.isna(dt):
        return pd.Series([None, None])
    lat, lon = location_info[plant]
    loc = pvlib.location.Location(lat, lon, tz='Asia/Seoul')
    pos = loc.get_solarposition(times=[dt])
    altitude = pos['apparent_elevation'].iloc[0]
    azimuth = pos['azimuth'].iloc[0]
    if altitude <= 0:
        azimuth = None
        altitude = 0
    return pd.Series([altitude, azimuth])

# 4) 계산 적용
df_1[['태양고도', '방위각']] = df_1.apply(compute_solar_pos, axis=1)

# 5) 시간대별 평균 고도 계산 (정상 확인용)
summary_hour = df_1.groupby(df_1['일시'].dt.hour)['태양고도'].mean().reset_index()
summary_hour.columns = ['시간대', '평균 고도 (°)']
print(summary_hour)
```
