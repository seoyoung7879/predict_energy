```python
import pandas as pd

# CSV íŒŒì¼ ì½ê¸°
df_all6 = pd.read_csv("df_all6.csv", encoding='utf-8')
df_forecast_updated = pd.read_csv("additional_farms.csv", encoding='utf-8')
```
