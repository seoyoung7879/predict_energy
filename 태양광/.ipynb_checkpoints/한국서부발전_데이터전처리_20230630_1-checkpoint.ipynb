```python
import pandas as pd
```

```python
df = pd.read_csv("C:\Users\kelly\OneDrive\바탕 화면\태양광\한국서부발전\한국서부발전(주)_태양광 발전 현황_20230630.csv", encoding = "cp949")
```

```python
#시간별 발전량 열 추출(3번 열~끝까지)
hourly_columns = df.columns[3:]

#melt 함수로 long-format으로 변환
hour_generationQuantity = df.melt(id_vars = df.columns[ :3],
                                 value_vars = hourly_columns,
                                 var_name = "시간",
                                 value_name = "해당 시간 발전량(kWh)")
```

```python
hour_generationQuantity
```

```python
hour_generationQuantity.to_csv("한국서부발전_시간별_발전량_utf-8.csv", index = False, encoding="utf-8-sig")
```
