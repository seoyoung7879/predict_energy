```python
df_list = [df감우리, df광양, df서귀포, df목포, df서산, df부산, df삼척소, df송당리, df수원, df신풍리, df영월, df울산, df이천, df익산, df인천, df하동]
df_통합 = pd.concat(df_list, ignore_index=True)
```

```python
# 중복 제거 (같은 행이 두 번 들어간 경우)
df_통합 = df_통합.drop_duplicates()
df_통합.head(3)
```

```python
df_통합["지점명"].unique()
```

```python
df_통합.info()
```

```python
df_통합["일시"].unique()
```

```python
# 1. 문자열 자르기: 초(:SS)가 있으면 잘라내고, 없으면 그대로 유지
df_통합["일시"] = df_통합["일시"].astype(str).str.slice(0, 16)  # 앞에서 16글자만 자름 → YYYY-MM-DD HH:MM
```

```python
# 2. 문자열을 datetime으로 변환
df_통합["일시"] = pd.to_datetime(df_통합["일시"], format="%Y-%m-%d %H:%M", errors="raise")
```

```python
df_통합["일시"].unique()
```

```python
df_통합["year"] = df_통합["일시"].dt.year
df_통합["month"] = df_통합["일시"].dt.month
df_통합["day"] = df_통합["일시"].dt.day
df_통합["weekday"] = df_통합["일시"].dt.weekday
df_통합["시간"] = df_통합["일시"].dt.hour

df_통합 = df_통합.drop(columns=["일시"])
```

```python
cols = ["지점명", "year", "month", "day", "weekday", "시간"]
df_통합 = df_통합[cols + ['기온(°C)', '강수량(mm)', '풍속(m/s)', '풍향(16방위)', '습도(%)', '현지기압(hPa)',
       '일조(hr)', '일사(MJ/m2)', '적설(cm)', '전운량(10분위)']]
df_통합.head(3)
```
