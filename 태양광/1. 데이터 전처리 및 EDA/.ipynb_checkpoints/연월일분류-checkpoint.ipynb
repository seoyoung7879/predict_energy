```python
import pandas as pd
```

```python
df = pd.read_csv("df_solar_real_final.csv", encoding = 'utf-8')
```

```python
df.head(3)
```

```python
df.info()
```

```python
# '일자' 컬럼을 datetime 형식으로 변환
df['일자'] = pd.to_datetime(df['일자'], errors='coerce')
```

```python
# 파생 변수 생성
df['year'] = df['일자'].dt.year
df['month'] = df['일자'].dt.month
df['day'] = df['일자'].dt.day
df['weekday'] = df['일자'].dt.weekday  # 0=월요일, 6=일요일
```

```python
# 결과 확인 (선택사항)
df[['일자', 'year', 'month', 'day', 'weekday']].head()
```

```python
df.head()
```

```python
#일자 삭제
df = df.drop(columns=["일자"])
```

```python
df.columns
```

```python
# 열 순서 재정렬
cols_priority = ['발전구분', 'year', 'month', 'day', 'weekday', '설비용량(MW)','시간', '발전량(kWh)', '연식(년)', '지점명', '기온(°C)',
       '강수량(mm)', '풍속(m/s)', '풍향(16방위)', '습도(%)', '일사(MJ/m2)', '적설(cm)',
       '하늘상태', '태양고도', '방위각']
df = df[cols_priority]
```

```python
# 결과 확인
df.head()
```

```python
df.to_csv("df_solar_real_final_yearmonthdaydate.csv", index=False, encoding="utf-8-sig")
```
