```python
# 발전량이 0인 경우 EDA 및 풍속이 높은데도 0인 경우 집중 분석
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib

# 한글 폰트 설정 (Windows: Malgun Gothic)
matplotlib.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False

# 데이터 불러오기
df = pd.read_csv('../../data/풍력_전처리_이상치제거.csv')

# 발전량이 0인 데이터만 추출
df_zero = df[df['발전량(kWh)'] == 0].copy()
print(f"발전량 0 데이터 개수: {len(df_zero)} / 전체: {len(df)} ({len(df_zero)/len(df)*100:.2f}%)")

# 발전량 0의 풍속 분포 시각화
plt.figure(figsize=(8,4))
sns.histplot(df_zero['풍속(m/s)'], bins=30, kde=True, color='skyblue')
plt.title('발전량 0인 경우의 풍속 분포')
plt.xlabel('풍속(m/s)')
plt.ylabel('Count')
plt.show()

# 발전량 0 vs 전체 풍속 분포 비교
plt.figure(figsize=(8,4))
sns.kdeplot(df['풍속(m/s)'], label='전체', color='gray')
sns.kdeplot(df_zero['풍속(m/s)'], label='발전량 0', color='red')
plt.title('전체 vs 발전량 0 풍속 분포')
plt.xlabel('풍속(m/s)')
plt.legend()
plt.show()

# 발전량 0인 경우의 주요 변수 통계
print('\n[발전량 0 데이터 주요 통계]')
print(df_zero[['풍속(m/s)', '기온(°C)', '강수량(mm)', '습도(%)', '적설(cm)', '태양고도']].describe())

# 발전량 0인 경우의 풍속별 개수 (구간별)
wind_bins = [0, 1, 2, 3, 4, 5, 6, 7, 8, 100]
df_zero['풍속구간'] = pd.cut(df_zero['풍속(m/s)'], bins=wind_bins)
print('\n[발전량 0 풍속 구간별 개수]')
print(df_zero['풍속구간'].value_counts().sort_index())

# 발전량 0 & 풍속 3m/s 이상인 경우만 따로 보기 (이상치 의심)
df_zero_highwind = df_zero[df_zero['풍속(m/s)'] >= 3]
print(f"\n풍속 3m/s 이상 & 발전량 0 데이터 개수: {len(df_zero_highwind)}")
if not df_zero_highwind.empty:
    display(df_zero_highwind[['일자','시간','지점명','풍속(m/s)','기온(°C)','강수량(mm)','태양고도']].head(10))

# 발전량 0의 시간대/지점/발전구분별 분포
plt.figure(figsize=(10,4))
sns.countplot(x='시간', data=df_zero, order=sorted(df_zero['시간'].unique()))
plt.title('발전량 0 시간대별 빈도')
plt.show()

plt.figure(figsize=(8,4))
sns.countplot(x='지점명', data=df_zero)
plt.title('발전량 0 지점별 빈도')
plt.show()

plt.figure(figsize=(8,4))
sns.countplot(x='발전구분', data=df_zero)
plt.title('발전량 0 발전구분별 빈도')
plt.show()

# 발전량 0의 일자별 빈도 (상위 10개)
if '일자' in df_zero.columns:
    top_dates = df_zero['일자'].value_counts().head(10)
    print('\n[발전량 0 상위 10일자]')
    print(top_dates)

# 발전량 0의 풍속-기온 산점도
plt.figure(figsize=(6,4))
sns.scatterplot(x='풍속(m/s)', y='기온(°C)', data=df_zero, alpha=0.3)
plt.title('발전량 0: 풍속 vs 기온')
plt.show()

# 팀원 의견 반영: 풍속이 충분히 높은데 발전량 0인 경우 집중 분석
print('\n[풍속 3m/s 이상 & 발전량 0 상세]')
if not df_zero_highwind.empty:
    display(df_zero_highwind[['일자','시간','지점명','풍속(m/s)','기온(°C)','강수량(mm)','태양고도']].describe())
    # 필요시 추가 EDA/시각화 가능
```
