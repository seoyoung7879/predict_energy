```python
# 오차 상위 10% 데이터 원인 분석 EDA
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
matplotlib.rc('font', family='Malgun Gothic')
matplotlib.rc('axes', unicode_minus=False)

# 데이터 로드
err_df = pd.read_csv('catboost_오차상위10p.csv')
print(f"데이터 shape: {err_df.shape}")

# 1. 오차 상위 케이스의 주요 변수 분포와 전체 분포 비교
print("\n=== 오차 상위 10%와 전체 분포 비교 ===")
try:
    full_df = pd.read_csv('파생변수추가_기상과풍력.csv')
    for col in ['풍속(m/s)', '기온(°C)', '습도(%)', '설비용량(MW)', '연식(년)']:
        if col in err_df.columns and col in full_df.columns:
            plt.figure(figsize=(10, 4))
            sns.kdeplot(full_df[col], label='전체', color='gray', fill=True, alpha=0.2)
            sns.kdeplot(err_df[col], label='오차상위10%', color='red', fill=True, alpha=0.4)
            plt.title(f'{col} 분포 (전체 vs 오차상위10%)')
            plt.legend()
            plt.tight_layout()
            plt.show()
except Exception as e:
    print(f"전체 데이터 분포 비교 불가: {e}")

# 2. 오차 상위 케이스의 결측/이상치/경계값 집중 여부
print("\n=== 오차 상위 10% 결측/이상치/경계값 집중 여부 ===")
for col in err_df.columns:
    if err_df[col].isnull().sum() > 0:
        print(f"{col} 결측치: {err_df[col].isnull().sum()}건")
    if err_df[col].dtype != 'object' and col not in ['실제','예측','오차','절대오차']:
        q1, q3 = err_df[col].quantile([0.25, 0.75])
        iqr = q3 - q1
        lower, upper = q1 - 1.5*iqr, q3 + 1.5*iqr
        outlier_cnt = ((err_df[col] < lower) | (err_df[col] > upper)).sum()
        if outlier_cnt > 0:
            print(f"{col} 이상치(1.5*IQR 밖): {outlier_cnt}건")

# 3. 발전구분/계절/호기 등 범주형 변수별 오차 집중
cat_cols = [col for col in ['발전구분','계절','호기'] if col in err_df.columns]
for col in cat_cols:
    print(f"\n=== {col}별 오차 상위 10% 분포 ===")
    print(err_df[col].value_counts())
    plt.figure(figsize=(8,4))
    sns.countplot(x=col, data=err_df)
    plt.title(f'오차상위10% {col}별 분포')
    plt.tight_layout()
    plt.show()

# 4. 주요 변수별 절대오차 상관관계 및 상위 변수별 분포
num_cols = [col for col in err_df.columns if err_df[col].dtype != 'object' and col not in ['실제','예측','오차','절대오차']]
if num_cols:
    corr = err_df[num_cols + ['절대오차']].corr()['절대오차'].sort_values(ascending=False)
    print("\n=== 절대오차와 주요 변수 상관관계 (상위 5) ===")
    print(corr.head(6))
    for col in corr.index[1:6]:
        plt.figure(figsize=(8,4))
        sns.scatterplot(x=err_df[col], y=err_df['절대오차'])
        plt.title(f'절대오차 vs {col}')
        plt.tight_layout()
        plt.show()

# 5. 실제-예측 차이(오차)가 큰 구간의 공통점 탐색
print("\n=== 오차 상위 10% 샘플 상위 10건 ===")
display_cols = ['실제','예측','오차','절대오차'] + [c for c in ['발전구분','계절','호기','풍속(m/s)','기온(°C)','습도(%)'] if c in err_df.columns]
print(err_df[display_cols].head(10))

# 6. 발전구분별 오차 상위 10%의 주요 변수 평균 비교
if '발전구분' in err_df.columns:
    print("\n=== 발전구분별 주요 변수 평균 (오차상위10%) ===")
    print(err_df.groupby('발전구분')[num_cols].mean())

print("\n[INFO] 오차 상위 10% 원인 탐색 EDA 완료")
```
